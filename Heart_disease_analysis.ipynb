# Analysis

```python
# Checking for outliers

df.plot(kind='box',subplots=True,figsize=(12,12),layout=(5,3))
```

**Insight : From the above plots we can see that there outliers in the datset**

```python
#Count of males and females

df['sex'].value_counts()
```

**Insight : There are 206 males and 96 females**

```python
# Displaying the count of the males and females

fig = sns.countplot(x='sex',data=df,palette="nipy_spectral_r")

for p in fig.patches:
    fig.annotate(p.get_height() , (p.get_x()+0.35, p.get_height()-60))
plt.title('Count of Males and Females', fontsize=20)
plt.xlabel('Sex', fontsize=15)
plt.ylabel('Count', fontsize=15)
```

**Insight : As the number of males is comparatively higher than the number of females, we may not get accurate results when we are doing gender comparisons with different features**  

```python
# Checking the rate of heat disease among the males and females with respect to age 

sns.catplot(data=df,x='sex',y='age',hue='target',palette='afmhot')
plt.title('Males and Females Diagnosed with Heart Disease', fontsize=20)
plt.xlabel('Sex', fontsize=15)
plt.ylabel('Age', fontsize=15)
```

**Insight : 
<br>â€¢ From the above plot it is understood that females are less likely to have heart disease than males
<br>â€¢ We can also see that chances of females getting a heart disease increases between the age of 55 and 65 years
<br>â€¢ And the chances of males getting a heart disease increases above 45 years of age**

```python
# Displaying the number of males and females having and not having heart disease

gen = pd.crosstab(df['sex'],df['target'])
print(gen)
gen.plot(kind='bar',stacked=True,grid=False)
plt.title('Male vs Females', fontsize=20)
plt.xlabel('Sex', fontsize=15)
plt.ylabel('Count', fontsize=15)
```

**Insight : From the dataset :
<br>â€¢ Females not having heart disease is 71 and having heart disease is 25
<br>â€¢ Males not having heart disease is 89 and having heart disease is 112**

```python
# Checking the target feature

plt.figure(figsize=(8,8))
labels=[""]
df['target'].value_counts().plot.pie(autopct="%.2f", startangle=90,textprops={'fontsize': 14})
plt.title('Heart Disease Population %', fontsize=20)
plt.ylabel('Target', fontsize=15)
```

<br>O - Heart Disease(absent)
<br>1 - Heart Disease(present)
<br>**Insight : From the dataset 46.13% of people have been diagnosed with heart disease**

```python
# Checking the chestpain type on the basis of target

plt.figure(figsize=(8,6))
plt.xticks(rotation=0, fontsize=18)
sns.countplot(x = df['target'],hue = 'chest_pain_type',data=df)
plt.xlabel("Target",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Chest Pain vs Target',fontsize=20)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

**Insight : Patients suffering from asymptomatic chest pain are more susceptible to having heart diseases
Asymptomatic Chest pain means shows no symptoms of heart disease, it is medically known as silent myocardial infarction**

```python
# Checking the chestpain type on the basis of sex

plt.figure(figsize=(8,6))
plt.xticks(rotation=0, fontsize=18)
sns.countplot(x=df['sex'], hue='chest_pain_type', data=df)
plt.title('Chest Pain Based On Gender', fontsize=20)
plt.xlabel('Sex', fontsize=15)
plt.ylabel('Counts', fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

**Insight : We can see that men are more susceptable from asymptomatic type of Chest Pain**

```python
# Checking the chestpain type on the basis of age

plt.figure(figsize=(24,8))
plt.xticks(rotation=0, fontsize=18)
sns.countplot(x=df['age'], hue='chest_pain_type', data=df)
plt.title('Chest Pain Based On Age', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Counts', fontsize=15)
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

**Insight : We can see that there is a increase in rate of heart disease between the ages of 50 and 65 and the most common type of chest pain is asymptomatic chest pain**

```python
# Comparing resting blood pressure and target 

plt.figure(figsize=(8,6))
sns.histplot(data=df,hue='target',x="resting_blood_pressure",multiple="dodge",palette="viridis")
plt.xlabel("Resting Blood Pressure",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Resting Blood Pressure VS Target',fontsize=20)
```

**Insight :<br>â€¢ Individuals with resting blood pressure less than 120 mm Hg have less chance of getting a heart disease
<br>â€¢ Resting blood pressure between 120 to 129 mm Hg indicates you have elevated blood pressure
<br>â€¢ Resting blood pressure between 130 â€“ 139 indicates you have high blood pressure(hypertension) stage 1
<br>â€¢ Resting blood pressure between 140 or higher indicates you have high blood pressure(hypertension) stage 2
<br>â€¢ Resting blood pressure higher than 180 indicates you have hypertensive crysis
<br> From the above plot we can see that blood pressure 140 or higher indicates that the target individual has heart disease**

```python
# Comparing fasting blood sugar and target

plt.figure(figsize=(8,6))
sns.countplot(data=df,hue='target',x="fasting_blood_sugar",palette="BuPu_r")
plt.xlabel("Fasting Blood Sugar",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Fasting Blood Sugar VS Target',fontsize=20)
plt.show()
```

**Insight : Patients having fasting blood suger >120 mg/dl have higher chances of having heart diseases**

```python
# Comparing fasting blood sugar and target

plt.figure(figsize=(12,6))
sns.histplot(data=df,hue='target',x="cholesterol",multiple="dodge",palette="dark:salmon_r")
plt.xlabel("Cholesterol Level",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Cholesterol VS Target',fontsize=20)
```

**Insight : Individuals having cholesterol level between 250 and above have a higher chances of getting  heart diseases**

```python
plt.figure(figsize=(8,6))
plt.xticks(rotation=0, fontsize=18)
sns.countplot(x='rest_ecg',hue='target',data=df,palette="Blues")
plt.xlabel("Resting Electrocardiographic Measurement",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Resting Electrocardiographic Measurement VS Target',fontsize=20)
plt.show()
```

```python
# Comparing heart rate and target

plt.figure(figsize=(8,6))
sns.histplot(x='max_heart_rate_achieved',hue='target',multiple='dodge',data=df,palette="vlag")
plt.xlabel("Maximum Heart Rate",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title('Maximum Heart Rate VS Target',fontsize=20)
```

```python
# Comparing exercise induced angina and target

plt.figure(figsize=(8,6))
plt.xticks(rotation=0, fontsize=18)
sns.countplot(x='exercise_induced_angina',data=df, hue='target',palette='cubehelix')
plt.xlabel("Exercise Induced Angina",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("Exercise Induced Angina VS target",fontsize=20)
plt.show
```

```python
# Comparing ST depression angina and target

plt.figure(figsize=(8,6))
sns.histplot(x='st_depression',hue='target',multiple='dodge',data=df,palette="icefire")
plt.xlabel("ST Depression",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.title("ST Depression VS target",fontsize=20)
```
